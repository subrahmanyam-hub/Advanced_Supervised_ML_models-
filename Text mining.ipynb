```python
elon=elon.drop(['Unnamed: 0'],axis=1)
```



## Data preprocessing 

```python
elon = [Text.strip() for Text in elon.Text] # remove both the leading and the trailing characters
elon = [Text for Text in elon if Text] # removes empty strings, because they are considered in Python as False
elon[0:10]
```

```python
text = ' '.join(elon)
text
```

```python
no_punc_text = text.translate(str.maketrans('', '', string.punctuation))
no_punc_text
```

```python
from nltk.tokenize import word_tokenize
text_tokens = word_tokenize(no_punc_text)
print(text_tokens[0:])
```

```python
len(text_tokens)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')

my_stop_words = stopwords.words('english')
my_stop_words.append('the')
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens[0:])
```

```python
lower_words = [Text.lower() for Text in no_stop_tokens]
print(lower_words[0:25])
```

### Lemmatization 

```python
nlp = spacy.blank('en') 
doc = nlp(' '.join(no_stop_tokens))
print(doc[0:])
```

```python
#Stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in lower_words]
print(stemmed_tokens[0:40])
```
